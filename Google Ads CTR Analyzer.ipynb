## Setting our imports and customization

```python
import pandas as pd
import re

pd.set_option('display.max_colwidth', -1)
pd.set_option('display.max_columns', 100)
```

## Importing Master DataFrame

```python
g_data = pd.DataFrame()
for i in range(1,6):
    csv_import = pd.read_csv('Google-Ads-EditorCurrent view+2019-07-31_part'+str(i)+'.csv')
    g_data = pd.concat([g_data, csv_import], sort=True)

g_data.reset_index(inplace=True)
g_data.sample(3)
```

## Cleaning Master DataFrame

```python
g_data.columns = g_data.columns.str.lower().str.replace(" ","_")
#g_data.drop(columns={'index'}, inplace=True)
g_data['ctr'] = g_data['ctr'].str.rstrip('%').astype('float')/100
g_data['ctr'] = g_data['ctr'].round(decimals=3)
g_data
```

## Creating CSV of Offers

```python
offers = g_data[g_data.labels.str.contains('(?i)offer', regex=True) == True]
offers = offers[offers.impressions > 200]
offers = offers[['account','impressions','avg_cpc','ctr','conv_rate','headline_1','headline_2','headline_3','description_line_1', 'description_line_2']]
offers.sort_values('ctr', ascending=False, inplace=True)

offers
```

```python
offers.to_csv('best_offer_ads.csv')
```

## Creating CSV of Branding
Just a simple copy/paste/replace of code above, no need to create a seperate function for this simple task.

```python
branding = g_data[(g_data.labels.str.contains('(?i)brand', regex=True) == True)&(g_data.campaign.str.contains('(?i)new', regex=True)==True)]
branding = branding[branding.impressions > 200]
branding = branding[['account','impressions','avg_cpc','ctr','conv_rate','headline_1','headline_2','headline_3','description_line_1', 'description_line_2']]
branding.sort_values('ctr', ascending=False, inplace=True)

branding
```

```python
branding.to_csv('best_branding_ads.csv')
```

## Creating CSVs of Ad Group Data
The definition below provides the ad information for [cars, suvs, & trucks] of the [preowned & general] campaigns

```python
#Use regEx keywords in def
def get_ad_data(ad_group, ad_campaign):
    temp_df = g_data[(g_data.ad_group.str.contains('(?i)'+ad_group+'', regex=True)==True)&(g_data.campaign.str.contains('(?i)'+ad_campaign+'', regex=True)==True)]
    temp_df = temp_df[temp_df.impressions > 200]
    temp_df = temp_df[['account','impressions','avg_cpc','ctr','conv_rate','headline_1','headline_2','headline_3','description_line_1', 'description_line_2']]
    temp_df.sort_values('ctr', ascending=False, inplace=True)
    temp_df.to_csv('best_'+ad_campaign+'_'+ad_group+'.csv')
    
for i in ['truck','suv','sedan|car']:
    get_ad_data(i,'preowned')
    get_ad_data(i,'general')
```
